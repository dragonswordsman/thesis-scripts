```python
from pathlib import Path
```

```python
config = ['HP-3fu-nofuse', 'HP-2fu-nofuse', 'HP-1fu-nofuse',
          'HP-3fu-fuse', 'HP-2fu-fuse', 'HP-1fu-fuse',
          'HP-3fu-fusepenalty', 'HP-2fu-fusepenalty', 'HP-1fu-fusepenalty',
          'LP-2fu-nofuse', 'LP-1fu-nofuse',
          'LP-2fu-fuse', 'LP-1fu-fuse',
          'LP-2fu-fusepenalty', 'LP-1fu-fusepenalty']
```

```python
benchmarks = {}
for c in config:
    folder = Path('stats/{}'.format(c))
    bench = list(folder.glob('**/stats.txt'))
    bench = {str(b.parent.relative_to(folder)) for b in bench}
    
    benchmarks[c] = bench
```

```python
with open('benchmarks.txt') as f:
    all_bench = f.readlines()
all_bench = {str(Path(l.rstrip())) for l in all_bench}
```

```python
for c in config:
    remain = list(all_bench.difference(benchmarks[c]))
    remain = [r[:-2] for r in remain]
    remain.sort(reverse=True)
    f = open('remain/{}.txt'.format(c.replace('-', '_')), 'w')
    f.write('\n'.join(map(lambda x: './{}.sh'.format(x), remain)))
    f.write('')
    f.close()
```
